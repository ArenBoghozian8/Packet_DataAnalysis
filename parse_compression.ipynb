```python
import csv
import matplotlib
import pandas as pd
import numpy as np
import datetime
import random
import matplotlib.pyplot as plt

import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.offline as py
import plotly.graph_objs as go

from ggplot import *
```

```python
def parse(ip, data):

    x = np.array(data[data['host_ip'] == ip]['test_date'])
    y = np.array(data[data['host_ip'] == ip]['discrimination_losses_str'])
    z = np.array(data[data['host_ip'] == ip]['base_losses_str'])

    x_ax = []
    y_ax = []
    z_ax = []
    diff = []
    for i in range(0, len(x)):
        x_ax.append(str(x[i][:][6:16]))
        y_ax.append(str((y[i]/5000) * 100))
        z_ax.append(str((z[i]/5000) * 100))
        diff.append( str(abs(((y[i]/5000) * 100) - ((z[i]/5000) * 100))))
        
    return x_ax, y_ax, z_ax, diff
```

```python
def draw(x_axis, y_axis, z_axis, diff, t, isSecondDay):
    
    #if isSecondDay:
    #    start = 14.5
    #    end = 38.5
    #else:
    #    start = 16.5
    #    end = 40.5
    
    trace1 = {"x":x_axis, "y":y_axis, "marker": {"color": "red", "size": 5, "symbol":"diamond"},
              "mode": "markers",
              "name": "Discriminated",
              "type": "scatter"#,
              #"mode":"lines+markers"
              }


    trace2 = {"x":x_axis, "y":z_axis, "marker": {"color": "green", "size": 5, "symbol":"square"},
              "mode": "markers",
              "name": "Base",
              "type": "scatter"#,
              #"mode":"lines+markers"
             }
    
    
    trace3 = {"x":x_axis, "y":diff, "marker": {"color": "blue", "size": 5},
              "mode": "markers",
              "name": "Difference",
              "type": "scatter"#,
              #"mode":"lines+markers"
             }    



    data = [trace1, trace2, trace3]

    layout = {"title": "Packet Loss " + t,
              "xaxis": {"title": "Time", },
              "yaxis": {"title": "Percent of Loss"}}

    layout = go.Layout(
        width = 2500,
        height = 2500 ,
        xaxis=dict(
            title = "Time",
            range=[5.5, 40.5]
        ),
        yaxis=dict(title="Percent of Loss"),

        title = 'Packet Loss ' + t,
    )

    fig = go.Figure(data=data, layout=layout)

    py.iplot(fig, filename = t ,image='png')
```

```python
data1 = pd.read_csv("Compression.csv")

ips = ['131.179.150.70','131.179.150.72', '192.16.125.12', '165.242.90.129', '130.195.4.68', '129.63.159.102', '192.91.235.230', '142.103.2.2']

ip_destination = ['planetlab1.cs.ucla.edu','planetlab2.cs.ucla.edu','planetlab-2.ssvl.kth.se', 'pl2.sos.info.hiroshima-cu.ac.jp', 'planetlab1.ecs.vuw.ac.nz', 'planetlab2.cs.uml.edu', 'pluto.cs.brown.edu', 'planetlab2.cs.ubc.ca']


for i in range(len(ips)):
    
    x ,y, z, diff = parse(ips[i], data1)
    draw(x,y,z,diff,ip_destination[i], isSecondDay = False)
 


#geom_point(position = 'identity', stat = 'identity')    
#p = ggplot(data1, aes('discrimination_losses_str', 'base_losses_str')) + \
#    geom_point()
    
#p.save('test.png')
    
#print(data1)
    
```
