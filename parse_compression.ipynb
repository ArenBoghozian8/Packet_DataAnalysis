```python
import csv
import matplotlib
import pandas as pd
import numpy as np
import datetime
import random
import matplotlib.pyplot as plt

import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.offline as py
import plotly.graph_objs as go
```

```python
def parse(ip, data):

    x = np.array(data[data['host_ip'] == ip]['test_date'])
    y = np.array(data[data['host_ip'] == ip]['discrimination_losses_str'])
    z = np.array(data[data['host_ip'] == ip]['base_losses_str'])

    x_ax = []
    y_ax = []
    z_ax = []
    for i in range(0, len(x), 2):
        x_ax.append(str(x[i][:][6:16]))
        y_ax.append(str((y[i]/5000) * 100))
        z_ax.append(str((z[i]/5000) * 100))
        
    return x_ax, y_ax, z_ax
```

```python
def draw(x_axis, y_axis, z_axis, t, d, isSecondDay):
    
    if isSecondDay:
        start = 14.5
        end = 38.5
    else:
        start = 16.5
        end = 40.5
    
    trace1 = {"x":x_axis, "y":y_axis, "marker": {"color": "red", "size": 12},
              "mode": "markers",
              "name": "Discriminated",
              "type": "scatter"}


    trace2 = {"x":x_axis, "y":z_axis, "marker": {"color": "green", "size": 12},
              "mode": "markers",
              "name": "Base",
              "type": "scatter"}



    data = [trace1, trace2]

    layout = {"title": "Packet Loss " + t,
              "xaxis": {"title": "Time", },
              "yaxis": {"title": "Percent of Loss"}}

    layout = go.Layout(
        width = 2500,
        height = 2500 ,
        xaxis=dict(
            title = "Time",
            #range=[start, end]
        ),
        yaxis=dict(title="Percent of Loss"),

        title = 'Packet Loss ' + t,
    )

    fig = go.Figure(data=data, layout=layout)

    py.iplot(fig, filename = t + '_' + d,image='png')
```

```python
data1 = pd.read_csv("CompressionAggregatesAllIPs.csv")
data2 = pd.read_csv("CompressionAggregatesGoodIPs.csv")

ips = ['165.242.90.129','131.179.150.70','129.63.159.102','192.91.235.230','142.103.2.2']

ip_destination = ['pl2.sos.info.hiroshima-cu.ac.jp','planetlab1.cs.ucla.edu','planetlab2.cs.uml.edu','pluto.cs.brown.edu','planetlab2.cs.ubc.ca']

for i in range(len(ips)):
    
    x ,y, z = parse(ips[i], data1)
    draw(x,y,z,ip_destination[i], d='day1', isSecondDay = False)
    
    x ,y, z = parse(ips[i], data2)
    draw(x,y,z,ip_destination[i], d='day2', isSecondDay = True)
    
```
